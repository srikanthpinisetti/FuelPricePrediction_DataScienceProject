#Data visualisations

```python
cols = clean.columns.to_list()
cols
```

```python
from matplotlib import markers
plt.figure(figsize=(45,10))
plt.plot(cols[0],cols[8],data=clean,marker='+',markeredgecolor='r')
plt.grid()
plt.title('Gasoline price trend')
plt.xlabel(cols[0])
plt.xticks(np.arange(0,len(clean)+1,12))
plt.ylabel(cols[8])
plt.savefig('Results/Visualizations/1. Gasoline_proce_trend.png')
plt.show()
```

## 2008 Great recession, Russo-Georgian war

------

```python
plt.figure(figsize=(35,25))
for i in range(1,8):
    plt.subplot(3,3,i)
    plt.grid()
    plt.scatter(cols[8],cols[i],data=clean)
    plt.xlabel(cols[8])
    plt.ylabel(cols[i])
plt.suptitle(f'Relationships of each independent features with {cols[8]}',fontsize=35)
plt.savefig('Results/Visualizations/2.Relationships.png')
plt.show()
```

##Correlation analysis

```python
corr = clean.corr()
corr.to_csv('Results/Stats/Correlation.csv')
corr
```

```python
plt.figure(figsize=(30,10))
plt.xticks(np.arange(0,len(clean)+1,12))
plt.yticks(np.arange(0,len(clean)+1,5))
plt.title(f'Trend of {cols[1]} and {cols[2]} over time')
plt.grid()
plt.plot(cols[0],cols[1],data=clean,label=cols[1])
plt.plot(cols[0],cols[2],data=clean,label=cols[2])
plt.legend()
plt.savefig('Results/Visualizations/3.Crude_oil trend.png')
plt.show()
```
